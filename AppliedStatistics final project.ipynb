```python
#we need to understand linearity between the dependent variable and independent variables

plt.scatter(df.N02BA, df.total_sales)

plt.title('N02BA vs. total_sales')
plt.xlabel('N02BA')
plt.ylabel('total_sales')
plt.show()
```

```python
import statsmodels.api as sm
x = df[[ 'M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03',
       'R06']]
y = df["total_sales"]

x = sm.add_constant(x) #adding a constant

model = sm.OLS(y, x).fit()
pred = model.predict(x,transform=False)
print_model = model.summary()
print(print_model)
```

```python
pred_y = model.predict(x)
pred_y = pd.Series(pred_y)
res = pd.DataFrame({'obs_y':y, 'pred_y':pred_y, 'Res':y-pred_y})
print(res.round(3))
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(3,3))
plt.hlines(y=0, xmin=df['M01AB'].min(),xmax=df['M01AB'].max(),linewidth=2)

plt.scatter(df['M01AB'], res['Res'],color = 'blue',s = 10)
plt.title("M01AB vs Residuals",fontsize=14)
plt.show()
```
